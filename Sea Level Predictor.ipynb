```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

    # Read data from file
df = pd.read_csv('epa-sea-level.csv', float_precision='legacy')

x = df['Year']
y = df['CSIRO Adjusted Sea Level']
```

```python
 # Create scatter plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(x,y, label='original data')
# Create first line of best fit
    
start_year = df['Year'].min() #df.iloc[0,df.columns.get_loc('Year')]
end_year = 2051
start_year2 = 2000

res = linregress(x,y)

x1 = list(range(start_year,end_year,1))
y1 = res.intercept + [res.slope * x1x for x1x in x1] 
    # Create second line of best fit
    
df_2000 = df[df.Year >= start_year2]

res2 = linregress(df_2000.Year, df_2000['CSIRO Adjusted Sea Level'])
x2 = list(range(start_year2,end_year,1))
y2 = res2.intercept + [res2.slope * x2x for x2x in x2]
    # Add labels and title
    
plt.plot(x1, y1 , 'orange', label='full fitted line >= 1880')
plt.plot(x2, y2, 'darkred', label='partial fitted line >= 2000')
    
ax.set_title('Rise in Sea Level')
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')  
  
plt.legend(loc='upper left')
plt.show()
  
```
